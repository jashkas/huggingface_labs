```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
import pandas as pd
from tabulate import tabulate

print("="*70)
print("СРАВНИТЕЛЬНЫЙ СЕНТИМЕНТ-АНАЛИЗ (2 МОДЕЛИ)")
print("="*70)

print("\n МОДЕЛЬ 1: RuBERT (Blanchefort/rubert-base-cased-sentiment)")
print("-" * 50)
model_name1 = "Blanchefort/rubert-base-cased-sentiment"
tokenizer1 = AutoTokenizer.from_pretrained(model_name1)
model1 = AutoModelForSequenceClassification.from_pretrained(model_name1)
sentiment_pipeline1 = pipeline("sentiment-analysis", model=model1, tokenizer=tokenizer1)
print("✓ Модель 1 загружена (метки: NEUTRAL, POSITIVE, NEGATIVE)")

print("\n МОДЕЛЬ 2: RuBERT (Tatyana/rubert-base-cased-sentiment-new)")
print("-" * 50)
model_name2 = "Tatyana/rubert-base-cased-sentiment-new"
tokenizer2 = AutoTokenizer.from_pretrained(model_name2)
model2 = AutoModelForSequenceClassification.from_pretrained(model_name2)
sentiment_pipeline2 = pipeline("sentiment-analysis", model=model2, tokenizer=tokenizer2)
print("✓ Модель 2 загружена (метки: NEUTRAL, POSITIVE, NEGATIVE)")

sentences = [
    "Этот фильм просто потрясающий, я в полном восторге!",
    "Прекрасная погода сегодня, отличный день для прогулки.",
    "Я получил повышение на работе, очень рад этому событию.",
    "Это лучший ресторан в городе, обслуживание на высшем уровне!",
    "Наконец-то я купил машину своей мечты, счастью нет предела!",
    "Это худший день в моей жизни, всё пошло не так.",
    "Обслуживание в этом магазине ужасное, больше никогда сюда не приду.",
    "Фильм оказался полным разочарованием, зря потратил время.",
    "Ужасная погода, целый день идет дождь и холодно.",
    "Мне нахамили в автобусе, очень неприятная ситуация.",
    "На улице идет дождь, температура воздуха +15 градусов.",
    "Завтра в 10 утра состоится совещание в конференц-зале.",
    "Средний чек в этом ресторане составляет 2000 рублей.",
    "Я прочитал эту книгу за три дня, она оказалась интересной.",
    "Поезд отправляется с третьего пути через 10 минут.",
    "Ну конечно, опять начальник отличился, молодец просто!",
    "Отличная работа, сломали всё, что можно было сломать.",
    "Какой замечательный день, чтобы остаться дома и никуда не выходить.",
    "Очень вкусно, особенно если сравнивать с обычной едой.",
    "Ну просто супер, всё как я люблю - опять ничего не работает."
]

expert_labels = [
    "POSITIVE", "POSITIVE", "POSITIVE", "POSITIVE", "POSITIVE",
    "NEGATIVE", "NEGATIVE", "NEGATIVE", "NEGATIVE", "NEGATIVE",
    "NEUTRAL", "NEUTRAL", "NEUTRAL", "NEUTRAL", "NEUTRAL",
    "NEGATIVE",
    "NEGATIVE",
    "NEUTRAL",
    "NEUTRAL",
    "NEGATIVE"
]

results = []

for i, sentence in enumerate(sentences):

    res1 = sentiment_pipeline1(sentence)[0]
    res2 = sentiment_pipeline2(sentence)[0]

    label1, score1 = res1['label'], res1['score']
    label2, score2 = res2['label'], res2['score']

    match1 = "Да" if label1 == expert_labels[i] else "Нет"
    match2 = "Да" if label2 == expert_labels[i] else "Нет"

    results.append({
        "№": i+1,
        "Предложение": sentence[:60] + "..." if len(sentence) > 60 else sentence,
        "Модель 1 (Blanchefort)": f"{label1} ({score1:.2f})",
        "Модель 2 (Tatyana)": f"{label2} ({score2:.2f})",
        "Эксперт": expert_labels[i],
        "Совпадение 1": match1,
        "Совпадение 2": match2
    })

df = pd.DataFrame(results)
print("\n" + "="*70)
print("РЕЗУЛЬТАТЫ СРАВНЕНИЯ")
print("="*70)
print(tabulate(df, headers='keys', tablefmt='grid', showindex=False))

accuracy1 = sum([1 for r in results if r["Совпадение 1"] == "Да"]) / len(results)
accuracy2 = sum([1 for r in results if r["Совпадение 2"] == "Да"]) / len(results)

print("\n" + "="*70)
print("СТАТИСТИКА КАЧЕСТВА")
print("="*70)
print(f"Модель 1 (Blanchefort) точность: {accuracy1*100:.1f}%")
print(f"Модель 2 (Tatyana) точность: {accuracy2*100:.1f}%")
```
