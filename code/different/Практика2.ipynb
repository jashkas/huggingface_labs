```python
# ЗАДАНИЕ 1
!pip install transformers torch --quiet

import torch
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification
from pprint import pprint
import pandas as pd

test_sentences = [
    "Основатель Tesla и SpaceX Илон Маск 10 апреля 2024 года посетил завод в Шанхае.",
    "В ходе визита в Москву 15 мая Генеральный секретарь ООН Антониу Гутерриш встретился с министром иностранных дел России Сергеем Лавровым.",
    "Саммит Большой двадцатки (G20) пройдет в Нью-Дели 9 и 10 сентября 2023 года, сообщает Reuters.",
    "Кандидат исторических наук Петров Сергей Иванович защитил диссертацию в МГУ имени М.В. Ломоносова 25 января 2025 года.",
    "Шведская компания IKEA объявила о возобновлении работы своего магазина в Киеве с 1 апреля 2024 года.",
    "На конференции в Париже 14 июля 1789 года никто не мог предсказать штурм Бастилии, но сегодня об этом знает каждый школьник."
]

print("-" * 50)
for i, sent in enumerate(test_sentences):
    print(f"{i+1}. {sent}")
print("-" * 50)

print(" Загрузка модели 1: Davlan/bert-base-multilingual-cased-ner-hrl...")
model_1_name = "Davlan/bert-base-multilingual-cased-ner-hrl"
tokenizer_1 = AutoTokenizer.from_pretrained(model_1_name)
model_1 = AutoModelForTokenClassification.from_pretrained(model_1_name)
ner_pipeline_1 = pipeline("ner", model=model_1, tokenizer=tokenizer_1, aggregation_strategy="simple")
print(" Модель 1 загружена.")

print(" Загрузка модели 2: arnolfokam/mbert-base-uncased-ner-pcm...")
model_2_name = "arnolfokam/mbert-base-uncased-ner-pcm"
tokenizer_2 = AutoTokenizer.from_pretrained(model_2_name)
model_2 = AutoModelForTokenClassification.from_pretrained(model_2_name)
ner_pipeline_2 = pipeline("ner", model=model_2, tokenizer=tokenizer_2, aggregation_strategy="simple")
print(" Модель 2 загружена.")
print("-" * 50)

def print_entities(sentence, results, model_title):
    """Выводит результаты NER в читаемом формате."""
    print(f"\n--- {model_title} ---")
    print(f"Предложение: {sentence}")
    if not results:
        print("  Сущностей не найдено.")
    else:
        for entity in results:

            entity_type = entity.get('entity_group', entity.get('entity', 'N/A'))

            if entity_type.startswith(('B-', 'I-')):
                entity_type = entity_type[2:]
            print(f"  - {entity['word']}: {entity_type} (уверенность: {entity['score']:.3f})")

results_summary = []

for i, sentence in enumerate(test_sentences):
    print(f"\n Анализ предложения {i+1}...")

    results_1 = ner_pipeline_1(sentence)
    print_entities(sentence, results_1, "Модель 1 (Davlan - PER/LOC/ORG)")

    results_2 = ner_pipeline_2(sentence)
    print_entities(sentence, results_2, "Модель 2 (arnolfokam - PER/LOC/ORG/DATE)")

    results_summary.append({
        "Предложение": sentence,
        "Модель 1 (Davlan)": [f"{e['word']} ({e.get('entity_group', e.get('entity', 'N/A')).replace('B-', '').replace('I-', '')})" for e in results_1],
        "Модель 2 (arnolfokam)": [f"{e['word']} ({e.get('entity_group', e.get('entity', 'N/A')).replace('B-', '').replace('I-', '')})" for e in results_2],
    })
    print("=" * 60)

print("\n\n ИТОГОВАЯ ТАБЛИЦА СРАВНЕНИЯ")
print("=" * 120)

df_results = pd.DataFrame(results_summary)
pd.set_option('display.max_colwidth', 200)
print(df_results.to_string(index=False))

print("\n" + "=" * 120)
```

```python
# ЗАДАНИЕ 2
!pip install transformers pandas torch --quiet

import pandas as pd
from transformers import pipeline
import torch
import gc

# Токен для доступа к моделям на Hugging Face
HF_TOKEN = ""

# Список моделей для тестирования
model_names = [
    "deepset/xlm-roberta-base-squad2",
    "mrm8488/bert-multi-cased-finetuned-xquadv1"
]

# Набор тестовых данных: контексты и вопросы к ним
qa_pairs = [
    {
        "context": "Байкал — самое глубокое озеро на Земле. Оно расположено в Восточной Сибири. Возраст озера оценивается примерно в 25 миллионов лет.",
        "questions": [
            "Где находится озеро Байкал?",
            "Какое озеро самое глубокое?",
            "Сколько лет озеру Байкал?"
        ]
    },
    {
        "context": "Александр Сергеевич Пушкин родился в Москве 6 июня 1799 года. Он считается основоположником современного русского литературного языка.",
        "questions": [
            "Когда родился Пушкин?",
            "В каком городе родился Пушкин?",
            "Кем считается Пушкин?"
        ]
    },
    {
        "context": "Солнце — звезда, которая находится в центре Солнечной системы. Земля вращается вокруг Солнца примерно за 365 дней.",
        "questions": [
            "Что находится в центре Солнечной системы?",
            "За сколько дней Земля вращается вокруг Солнца?"
        ]
    }
]

# Список для сбора всех результатов
all_results = []

# Параметры по умолчанию (базовые)
base_params = {
    "name": "Базовые",
    "top_k": 1,
    "doc_stride": 128,
    "max_answer_len": 50,
    "max_seq_len": 384,
    "max_question_len": 64
}

def test_model(model_name, param_dict, qa_pairs):
    """Функция для тестирования одной модели с заданными параметрами."""
    print(f"\nМодель: {model_name} | Параметры: {param_dict['name']}")
    try:
        # Инициализация пайплайна для вопросно-ответной задачи
        qa = pipeline(
            "question-answering",
            model=model_name,
            token=HF_TOKEN,
            device=0 if torch.cuda.is_available() else -1
        )

        # Перебор всех контекстов и вопросов
        for pair in qa_pairs:
            context = pair["context"]
            for question in pair["questions"]:
                result = qa(
                    question=question,
                    context=context,
                    top_k=param_dict.get("top_k", 1),
                    doc_stride=param_dict.get("doc_stride", 128),
                    max_answer_len=param_dict.get("max_answer_len", 50),
                    max_seq_len=param_dict.get("max_seq_len", 384),
                    max_question_len=param_dict.get("max_question_len", 64)
                )
                # Если вернулся список ответов (top_k > 1), обрабатываем каждый
                if isinstance(result, list):
                    for i, ans in enumerate(result):
                        all_results.append({
                            "Модель": model_name,
                            "Параметры": param_dict['name'],
                            "Контекст": context[:60] + "..." if len(context) > 60 else context,
                            "Вопрос": question,
                            "Ответ": ans['answer'],
                            "Уверенность": f"{ans['score']:.2%}",
                            "Позиция": f"[{ans['start']}:{ans['end']}]",
                            "Ранг": i+1
                        })
                else:
                    all_results.append({
                        "Модель": model_name,
                        "Параметры": param_dict['name'],
                        "Контекст": context[:60] + "..." if len(context) > 60 else context,
                        "Вопрос": question,
                        "Ответ": result['answer'],
                        "Уверенность": f"{result['score']:.2%}",
                        "Позиция": f"[{result['start']}:{result['end']}]",
                        "Ранг": 1
                    })
        # Освобождение ресурсов
        del qa
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        return True
    except Exception as e:
        print(f"Ошибка при загрузке/использовании модели {model_name}: {e}")
        return False

# Запуск тестирования для всех моделей с базовыми параметрами
print("="*70)
print("ЗАДАНИЕ 2: ТЕСТИРОВАНИЕ ДВУХ МОДЕЛЕЙ (БАЗОВЫЕ ПАРАМЕТРЫ)")
print("="*70)

for model in model_names:
    test_model(model, base_params, qa_pairs)

# Формирование итоговой таблицы для задания 2
if all_results:
    df_all = pd.DataFrame(all_results)
    df_task2 = df_all[df_all['Параметры'] == "Базовые"].copy()
    print("\n\nТАБЛИЦА ДЛЯ ЗАДАНИЯ 2")
    # Группировка по моделям для наглядности
    for model in df_task2['Модель'].unique():
        print(f"\n{model}")
        model_df = df_task2[df_task2['Модель'] == model][['Вопрос', 'Ответ', 'Уверенность']]
        print(model_df.to_string(index=False))
else:
    print("Нет данных для отображения.")
```

```python

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
import time
import pandas as pd
from tabulate import tabulate
import re

print("="*70)
print("ЗАДАНИЕ 5: МОДИФИКАЦИЯ КОДА ДЛЯ РАЗНЫХ МОДЕЛЕЙ")
print("="*70)

def generate_summary(model, tokenizer, text, **kwargs):
    """Функция для генерации суммаризации с заданными параметрами"""
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=1024)

    with torch.no_grad():
        outputs = model.generate(
            inputs["input_ids"],
            **kwargs
        )

    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary

def evaluate_summary(summary, original_text, tokenizer):
    """Оценивает качество суммаризации по разным метрикам"""
    summary_len = len(tokenizer.encode(summary))
    original_len = len(tokenizer.encode(original_text))

    ends_with_dot = 1 if summary.strip().endswith(('.', '!', '?')) else 0

    key_phrases = ['Байкал', 'озеро', 'вода', 'глубина', 'Россия', 'эндемик', 'нерпа']
    key_words_found = sum(1 for phrase in key_phrases if phrase.lower() in summary.lower())

    sentences = summary.split('.')
    has_structure = 1 if len([s for s in sentences if len(s.strip()) > 10]) > 1 else 0

    quality_score = (
        (30 * (1 - abs(summary_len - 70) / 70)) +
        (20 * ends_with_dot) +
        (25 * min(key_words_found / 3, 1)) +
        (25 * has_structure)
    )

    return {
        'length': summary_len,
        'compression_ratio': summary_len / original_len * 100,
        'ends_with_dot': ends_with_dot,
        'key_words_found': key_words_found,
        'has_structure': has_structure,
        'quality_score': round(quality_score, 2)
    }

print("\n1. Загружаем модель cointegrated/rut5-base-absum...")
model_name1 = "cointegrated/rut5-base-absum"
tokenizer1 = AutoTokenizer.from_pretrained(model_name1)
model1 = AutoModelForSeq2SeqLM.from_pretrained(model_name1)
print("Модель 1 загружена")

print("2. Загружаем модель cointegrated/rut5-base-multitask...")
model_name2 = "cointegrated/rut5-base-multitask"
tokenizer2 = AutoTokenizer.from_pretrained(model_name2)
model2 = AutoModelForSeq2SeqLM.from_pretrained(model_name2)
print("Модель 2 загружена")

print("3. Загружаем модель IlyaGusev/mbart_ru_sum_gazeta...")
model_name3 = "IlyaGusev/mbart_ru_sum_gazeta"
tokenizer3 = AutoTokenizer.from_pretrained(model_name3)
model3 = AutoModelForSeq2SeqLM.from_pretrained(model_name3)
print("Модель 3 загружена")

print("\nВсе модели успешно загружены!")

text = """
Озеро Байкал — это уникальное природное образование, расположенное в Восточной Сибири на территории
России. Оно является самым глубоким озером на планете, его максимальная глубина достигает 1642 метра.
Байкал содержит около 20% всех мировых запасов пресной воды — это примерно 23 615 кубических километров,
что превышает объем всех пяти Великих озер Северной Америки вместе взятых.

Возраст озера оценивается учеными в 25-30 миллионов лет, что делает его одним из древнейших озер в мире.
В Байкал впадает более 300 рек и ручьев, а вытекает только одна река — Ангара. Крупнейшие притоки —
Селенга, Верхняя Ангара, Баргузин. Уровень воды в озере регулируется естественным путем и зависит
от климатических условий.

Байкал славится уникальным животным и растительным миром. Более 80% видов живых организмов,
обитающих в озере, являются эндемиками, то есть не встречаются больше нигде в мире. Самый известный
эндемик — байкальская нерпа, единственный в мире вид тюленя, обитающий в пресной воде. Также
в озере водится знаменитый байкальский омуль, голомянка — прозрачная рыба без чешуи, и множество
других уникальных видов.

Вода в Байкале отличается исключительной чистотой и прозрачностью. Весной можно видеть предметы
на глубине до 40 метров. Это объясняется наличием микроскопических рачков эпишуры, которые фильтруют
воду, пропуская ее через себя. Вода содержит очень мало минеральных веществ и близка по составу
к дистиллированной.
"""

inputs = tokenizer1(text, return_tensors="pt", truncation=True, max_length=1024)
original_len = inputs["input_ids"].shape[1]
print(f"\nДлина исходного текста: {original_len} токенов")

print("\n" + "="*70)
print("ЗАДАНИЕ 4: ЭКСПЕРИМЕНТЫ С ПАРАМЕТРАМИ СУММАРИЗАЦИИ")
print("="*70)


param_sets = [
    {
        "name": "Базовые (исправленные)",
        "max_length": 150,
        "min_length": 50,
        "num_beams": 4,
        "do_sample": False,
        "temperature": 1.0,
        "repetition_penalty": 2.5,
        "length_penalty": 1.5,
        "no_repeat_ngram_size": 3,
        "early_stopping": True,
        "num_return_sequences": 1
    },
    {
        "name": "Короткие предложения",
        "max_length": 80,
        "min_length": 30,
        "num_beams": 4,
        "do_sample": False,
        "temperature": 1.0,
        "repetition_penalty": 3.0,
        "length_penalty": 1.2,
        "no_repeat_ngram_size": 4,
        "early_stopping": True
    },
    {
        "name": " Полные предложения",
        "max_length": 200,
        "min_length": 80,
        "num_beams": 5,
        "do_sample": False,
        "temperature": 1.0,
        "repetition_penalty": 2.0,
        "length_penalty": 1.0,
        "no_repeat_ngram_size": 3,
        "early_stopping": True
    },
    {
        "name": "Творческий режим",
        "max_length": 150,
        "min_length": 50,
        "num_beams": 1,
        "do_sample": True,
        "temperature": 0.9,
        "repetition_penalty": 2.2,
        "length_penalty": 1.3,
        "no_repeat_ngram_size": 3,
        "early_stopping": False,
        "top_p": 0.9,
        "top_k": 50
    },
    {
        "name": "Быстрый режим",
        "max_length": 120,
        "min_length": 40,
        "num_beams": 2,
        "do_sample": False,
        "temperature": 1.0,
        "repetition_penalty": 2.5,
        "length_penalty": 1.5,
        "no_repeat_ngram_size": 3,
        "early_stopping": True
    },
    {
        "name": " Оптимальный баланс",
        "max_length": 150,
        "min_length": 60,
        "num_beams": 4,
        "do_sample": False,
        "temperature": 1.0,
        "repetition_penalty": 2.2,
        "length_penalty": 1.3,
        "no_repeat_ngram_size": 3,
        "early_stopping": True
    }
]


models = [
    {"name": "rut5-base-absum", "model": model1, "tokenizer": tokenizer1},
    {"name": "rut5-base-multitask", "model": model2, "tokenizer": tokenizer2},
    {"name": "mbart_ru_sum_gazeta", "model": model3, "tokenizer": tokenizer3}
]


results = []

print("\nТЕСТИРОВАНИЕ МОДЕЛЕЙ С РАЗНЫМИ ПАРАМЕТРАМИ:")
print("-" * 80)

for model_info in models:
    print(f"\nМОДЕЛЬ: {model_info['name']}")
    print("-" * 40)

    for params in param_sets:
        print(f"\n  {params['name']}")

        try:
            start_time = time.time()


            generate_params = {
                'max_length': params['max_length'],
                'min_length': params['min_length'],
                'num_beams': params['num_beams'],
                'do_sample': params['do_sample'],
                'repetition_penalty': params['repetition_penalty'],
                'length_penalty': params['length_penalty'],
                'no_repeat_ngram_size': params['no_repeat_ngram_size'],
                'early_stopping': params['early_stopping']
            }


            if params['do_sample']:
                generate_params['temperature'] = params.get('temperature', 1.0)
                if 'top_p' in params:
                    generate_params['top_p'] = params['top_p']
                if 'top_k' in params:
                    generate_params['top_k'] = params['top_k']


            summary = generate_summary(
                model_info['model'],
                model_info['tokenizer'],
                text,
                **generate_params
            )

            elapsed_time = time.time() - start_time


            evaluation = evaluate_summary(summary, text, model_info['tokenizer'])

            print(f"    Результат: {summary[:150]}...")
            print(f"    Длина: {evaluation['length']} токенов | Сжатие: {evaluation['compression_ratio']:.1f}%")
            print(f"    Качество: {evaluation['quality_score']}/100 | Время: {elapsed_time:.2f} сек")
            print(f"    Завершено: {'!' if evaluation['ends_with_dot'] else '✗'} | Ключевых слов: {evaluation['key_words_found']}/7")


            results.append({
                'Модель': model_info['name'],
                'Параметры': params['name'],
                'Длина_токены': evaluation['length'],
                'Сжатие_%': round(evaluation['compression_ratio'], 1),
                'Качество': evaluation['quality_score'],
                'Время_сек': round(elapsed_time, 2),
                'Завершено': '✓' if evaluation['ends_with_dot'] else '✗',
                'Ключевые_слова': evaluation['key_words_found'],
                'Текст': summary
            })

        except Exception as e:
            print(f"Ошибка: {str(e)[:100]}")


print("\n" + "="*70)
print("АНАЛИЗ РЕЗУЛЬТАТОВ")
print("="*70)

if results:
    df = pd.DataFrame(results)


    print("\nСВОДНАЯ ТАБЛИЦА ПО МОДЕЛЯМ:")
    model_summary = df.groupby('Модель').agg({
        'Качество': 'mean',
        'Время_сек': 'mean',
        'Длина_токены': 'mean',
        'Завершено': lambda x: sum(1 for v in x if v == '✓') / len(x) * 100
    }).round(2)
    model_summary.columns = ['Ср. качество', 'Ср. время (сек)', 'Ср. длина', '% завершенных']
    print(tabulate(model_summary, headers='keys', tablefmt='grid'))


    print("\nВЛИЯНИЕ ПАРАМЕТРОВ НА КАЧЕСТВО:")
    param_summary = df.groupby('Параметры').agg({
        'Качество': ['mean', 'max'],
        'Время_сек': 'mean',
        'Завершено': lambda x: sum(1 for v in x if v == '✓') / len(x) * 100
    }).round(2)
    print(tabulate(param_summary, headers='keys', tablefmt='grid'))


    print("\nТОП-5 ЛУЧШИХ РЕЗУЛЬТАТОВ:")
    top_results = df.nlargest(5, 'Качество')[['Модель', 'Параметры', 'Качество', 'Длина_токены', 'Время_сек']]
    print(tabulate(top_results, headers='keys', tablefmt='grid', showindex=False))


print("\n" + "="*70)
print("АНАЛИЗ ПРОБЛЕМЫ ОБРЫВА ПРЕДЛОЖЕНИЙ")
print("="*70)

print("""
ПРИЧИНЫ ОБРЫВА ПРЕДЛОЖЕНИЙ:
1. Слишком маленький max_length не дает закончить предложение
2. Высокий length_penalty (>2.0) слишком сильно поощряет короткие тексты
3. Недостаточный min_length позволяет остановиться на полуслове

РЕШЕНИЯ:
1. Увеличить min_length до 50-60 токенов
2. Уменьшить length_penalty до 1.0-1.5
3. Использовать early_stopping=True с достаточным max_length
4. Увеличить num_beams до 4-5 для более связного текста
""")

print("\n" + "="*70)
print("ЛУЧШИЕ РЕЗУЛЬТАТЫ С ИСПРАВЛЕННЫМИ ПАРАМЕТРАМИ")
print("="*70)

for model_name in ["rut5-base-absum", "rut5-base-multitask", "mbart_ru_sum_gazeta"]:
    model_results = [r for r in results if r['Модель'] == model_name]
    if model_results:
        best = max(model_results, key=lambda x: x['Качество'])
        print(f"\n{model_name} (качество: {best['Качество']}/100):")
        print(f"   Параметры: {best['Параметры']}")
        print(f"   Результат: {best['Текст']}")
        print(f"   Длина: {best['Длина_токены']} токенов | Время: {best['Время_сек']} сек")
        print(f"   {'Завершено' if best['Завершено'] == '✓' else '✗ Обрыв'}")

```

```python

# !pip install --upgrade transformers

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Загружаем модель перевода с Hugging Face Hub
print("Загружаем модель для бурятского...")

MODEL_NAME = "SaranaAbidueva/nllb-200-bxr-ru"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)

def translate(text):
    # ШАГ 1: Токенизация - превращаем текст в числа (тензоры)
    inputs = tokenizer.encode(text, return_tensors="pt")

    # ШАГ 2: Генерация - нейросеть создает перевод в виде чисел
    outputs = model.generate(inputs)

    # ШАГ 3: Декодирование - превращаем числа обратно в текст
    # decode() преобразует числа в слова, skip_special_tokens=True убирает служебные символы
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

```

```python
# Используем нашу функцию для перевода
# Пример текста на русском, который хотим перевести на бурятский
text = "Доброе утро!"
# Или бурятского на русский:
# text = "Сайн байна, ямар боломж?"

result = translate(text)

# Выводим результаты
print(f"🇷🇺 Русский: {text}")
print(f"🇧🇷 Бурятский: {result}")
```
