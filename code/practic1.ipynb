```python
from transformers import pipeline
```

```python
texts = [
    "Этот фильм просто великолепен!",
    "Я очень доволен результатом.",
    "Спасибо за отличную работу!",
    "Ужасное обслуживание, больше не приду.",
    "Товар оказался бракованным.",
    "Фильм скучный и затянутый.",
    "На улице идёт дождь."
]

expert_labels = ["POSITIVE", "POSITIVE", "POSITIVE", "NEGATIVE", "NEGATIVE", "NEGATIVE", "NEUTRAL"]

# Для сбора статистики
total = len(texts)
```

```python
# Создаём глобальный словарь для хранения результатов
results = {}

def sentiment_analysis(model_id, model_name, texts=texts, expert_labels=expert_labels, results_dict=None):
    if results_dict is None:
        results_dict = results  # используем глобальный словарь по умолчанию

    classifier = pipeline("sentiment-analysis", model=model_name)
    
    print(f"\nМодель {model_id}: {model_name}")
    print("-" * 100)
    correct = 0
    total = len(texts)
    
    for text, expert in zip(texts, expert_labels):
        result = classifier(text)[0]
        model_label = result['label']
        score = result['score']
        match = 1 if model_label.upper() == expert else 0
        correct += match
        print(f"Текст: {text[:30]}... | Эксперт: {expert} | Модель: {model_label} ({score:.4f}) | Совпадение: {match}")
    
    percent = (correct / total) * 100
    # Кладём результат в словарь по ключу model_name (старое значение перезапишется)
    results_dict[model_name] = percent
    print(f"| Итоговый % совпадений модели с мнением эксперта {percent:.1f}")
```

```python
from transformers import pipeline
import pandas as pd
```

```python
def test_zero_shot(model_name, texts, candidate_labels, expert_labels, model_id=None):
    print(f"\n--- Модель {model_id}: {model_name} ---")
    classifier = pipeline("zero-shot-classification", model=model_name)
    correct = 0
    total = len(texts)
    for i, (text, expert) in enumerate(zip(texts, expert_labels)):
        result = classifier(text, candidate_labels)
        predicted = result['labels'][0]
        score = result['scores'][0]
        match = 1 if predicted == expert else 0
        correct += match
        print(f"{i+1}. {text[:50]:<50} | Эксперт: {expert:<8} | Модель: {predicted} ({score:.2%}) | Совпадение: {match}")
    percent = correct / total * 100
    print(f"ИТОГ: {percent:.1f}% совпадений")
    return percent
```

```python
def explore_parameters(model_name, text, labels):
    print(f"\n=== Модель: {model_name} ===")
    classifier = pipeline("zero-shot-classification", model=model_name)

    # 1. Без параметров
    res = classifier(text, labels)
    print("1. Без параметров:", res['labels'][0], f"({res['scores'][0]:.2%})")

    # 2. hypothesis_template
    res = classifier(text, labels, hypothesis_template="Этот текст относится к теме {}.")
    print("2. С hypothesis_template:", res['labels'][0], f"({res['scores'][0]:.2%})")

    # 3. multi_label=True (покажет вероятности для всех классов)
    res = classifier(text, labels, multi_label=True)
    print("3. multi_label=True:", dict(zip(res['labels'], res['scores'])))

    # 4. batch_size (для одного текста не влияет, но покажем)
    res = classifier(text, labels, batch_size=2)
    print("4. batch_size=2:", res['labels'][0], f"({res['scores'][0]:.2%})")
```

```python
model_name = "cointegrated/rubert-base-cased-nli-threeway"
text = "Футбольный матч закончился со счётом 3:1 в пользу хозяев."
labels = ["спорт", "политика", "наука"]
explore_parameters(model_name, text, labels)
```

```python
model_name = "brahmairesearch/TFS-ZS-MNLI"
text = "Футбольный матч закончился со счётом 3:1 в пользу хозяев."
labels = ["спорт", "политика", "наука"]
explore_parameters(model_name, text, labels)
```

```python
from transformers import pipeline

prompts = [
    "В этом курсе мы научим вас, как",
    "Однажды в лесу"
]

# Словарь для накопления субъективных оценок
results_gen = {}
```

```python
def test_generation(model_name, prompts, model_id=None, **gen_kwargs):
    print(f"\n--- Модель {model_id}: {model_name} ---")
    try:
        generator = pipeline("text-generation", model=model_name)
    except Exception as e:
        print(f"Не удалось загрузить модель: {e}")
        return None

    outputs = []
    for i, prompt in enumerate(prompts):
        try:
            results = generator(prompt, **gen_kwargs)
            print(f"Промпт {i+1}: {prompt}")
            generated = []
            for j, res in enumerate(results):
                text = res['generated_text']
                print(f"  {j+1}. {text}")
                generated.append(text)
            outputs.append(generated)
        except Exception as e:
            print(f"Ошибка при генерации для промпта {i+1}: {e}")
            outputs.append(None)
    return outputs  # возвращаем список сгенерированных текстов
```

```python
model_name = "ai-forever/rugpt3small_based_on_gpt2"
out = test_generation(model_name, prompts, model_id=1, max_new_tokens=30, num_return_sequences=2, do_sample=True, temperature=0.8)
if out is not None:
    results_gen[model_name] = out
```

```python
model_name = "ai-forever/rugpt3medium_based_on_gpt2"
rating = test_generation(model_name, prompts, model_id=2, max_new_tokens=50, num_return_sequences=2, do_sample=True, temperature=0.8)
if rating:
    results_gen[model_name] = rating
```

```python
model_name = "ai-forever/rugpt3large_based_on_gpt2"
rating = test_generation(model_name, prompts, model_id=3, max_new_tokens=30, num_return_sequences=2, do_sample=True, temperature=0.8)
if rating:
    results_gen[model_name] = rating
```

```python
model_name = "bigscience/bloom-1b1"
out = test_generation(model_name, prompts, model_id=4, max_new_tokens=30, num_return_sequences=2, do_sample=True, temperature=0.8)
if out is not None:
    results_gen[model_name] = out
```

### Задание 7
Найдите 5 моделей, работающих с русским языком. Поэкспериментируйте с ними. Сделайте вывод о качестве каждой модели в виде таблицы.

```python
# Предложения с маской
masked_texts = [
    "Ребенок мыл [MASK].",
    "На столе стояла [MASK] с цветами.",
    "Он работает [MASK] в крупной компании."
]

results_mask = {}

def test_fill_mask(model_name, texts, model_id=None, top_k=5):
    print(f"\n--- Модель {model_id}: {model_name} ---")
    try:
        unmasker = pipeline("fill-mask", model=model_name)
    except Exception as e:
        print(f"Не удалось загрузить модель: {e}")
        return None

    outputs = []
    for text in texts:
        try:
            results = unmasker(text, top_k=top_k)
            print(f"Текст: {text}")
            for r in results:
                print(f"  {r['token_str']} (score={r['score']:.4f})")
            outputs.append(results)
        except Exception as e:
            print(f"Ошибка при обработке текста '{text}': {e}")
            outputs.append(None)
    return outputs
```

```python
# Модель 1
model_name = "DmitryPogrebnoy/distilbert-base-russian-cased"
out = test_fill_mask(model_name, masked_texts, model_id=1, top_k=5)
if out:
    results_mask[model_name] = out
```

```python
# Модель 2
model_name = "DeepPavlov/rubert-base-cased"
out = test_fill_mask(model_name, masked_texts, model_id=2, top_k=5)
if out:
    results_mask[model_name] = out
```

```python
# Модель 3
model_name = "sberbank-ai/ruBert-base"
out = test_fill_mask(model_name, masked_texts, model_id=3, top_k=5)
if out:
    results_mask[model_name] = out
```

```python
# Модель 4
model_name = "cointegrated/rubert-base-cased-nli-threeway"
out = test_fill_mask(model_name, masked_texts, model_id=4, top_k=5)
if out:
    results_mask[model_name] = out
```

```python
# Итоговая таблица
print("\n" + "="*80)
print("ИТОГОВАЯ ТАБЛИЦА (Fill-mask)")
print("="*80)
print(f"{'Название модели':60} {'Лучший вариант для первого текста':30}")
print("-" * 90)

for model, outputs in results_mask.items():
    if outputs and outputs[0]:
        best = outputs[0][0]['token_str'] if outputs[0] else "Ошибка"
        print(f"{model[:60]:60} {best:30}")
    else:
        print(f"{model[:60]:60} {'Ошибка':30}")

print("\nВывод: модели на базе BERT показывают осмысленные предсказания. DeepPavlov/rubert-base-cased работают хорошо.")
```
